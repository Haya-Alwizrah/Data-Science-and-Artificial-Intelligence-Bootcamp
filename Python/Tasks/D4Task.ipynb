## Task 1: Check Divisibility
Ask the user to enter a number. If the number is divisible by both 3 and 5, print `FizzBuzz`.

```python
#write your solution
x = int(input())
if x %3 ==0 & x%5 ==0:
    print("FizzBuzz")
```

## Task 3: Voting Eligibility with Citizenship
Ask the user to enter their age and whether they are a citizen (`yes` or `no`).
Print `Eligible` if age is 18 or older and citizen is `yes`.

```python
#write your solution
x = int(input("enter u age"))
y = input("citized? Yes, No").lower()

citizen = True if y == "yes" else False

if x>=18 or citizen:
    print("eligible")
```

## Task 4: Compare Two Numbers
Ask the user to enter two numbers and print:
- `Equal` if they are the same
- `First is greater` if the first is greater
- `Second is greater` otherwise

```python
#write your solution

x = int(input())
y = int(input())

if x ==y:
    print("equal")
elif x>y:
    print("first greater")
else:
    print("second greater")
```

## Task 5: Grade with Bonus
Ask for a score and whether the student has a bonus (`yes` or `no`).
If `yes`, add 5 points. Then display the grade:
- 90+: Excellent
- 75–89: Good
- 60–74: Pass
- Below 60: Fail

```python
#write your solution

x = int(input())
y = input("have bonus? Yes, No").lower()

bonus = True if y == "yes" else False

if bonus:
    x += 5

if x >= 90:
    print("excellent")
elif x >= 75:
    print("good")
elif x >= 60:
    print("pass")
else:
    print("fail")
```

## Task 7: Temperature Advice
Ask for the temperature in Celsius and print:
- `Very hot` if > 40
- `Cold` if < 10
- `Perfect weather` if between 20 and 30
- `Normal` otherwise

```python
#write your solution

x = int(input())

if x > 40:
    print("very hot")
elif x < 10:
    print("normal")
elif 20 < x < 30:
    print("perfect")
else:
    print("normal")
```

## Task 8: Complex Condition
Ask for age and GPA. If age < 25 or GPA > 4.5, print `You are eligible for the program`. Otherwise, print `Not eligible`.

```python
#write your solution

x = int(input())
y = float(input())

if (x < 25) | (y>4.5):
    print("You are eligible for the program")
else:
    print("Not eligible")
```

## Task 9: Arithmetic Checker
Ask for two numbers and an operator (`+`, `-`, `*`, `/`). Perform the operation using `if-elif` conditions.

```python
#write your solution

x = int(input())
y = int(input())

z = input("chose operator:(+, -, *, /)")

if z == "+":
    print(x+y)
elif z == "-":
    print(x-y)
elif z == "*":
    print(x*y)
elif z == "/":
    print(x/y)
else:
    print("error")
```

## Task 10: Ternary Challenge
Ask the user to enter a number. Print `Multiple of 10` if divisible by 10, else print `Not a multiple of 10` using a ternary expression.

```python
#write your solution

x = int(input())

print("Multiple of 10") if x %10 ==0 else print("Not a multiple of 10")
```
