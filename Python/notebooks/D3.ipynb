```python
wallet = 9
juice = 4
chips = 6

x = int(input("enter u chose: 1,2,3"))

def machine(x):
    if x == 1:
        total = juice
    elif x == 2:
        total = chips
    elif x == 3:
        total = juice + chips
    else:
        print("invalid choice")
        return

    if wallet < total:
        print("u didnt have enouph money")
    elif total == 0:
        print("invalid choice")
    else:
        print(f"thank you your total {total}, u have {wallet-total} left")


machine(x)
```
