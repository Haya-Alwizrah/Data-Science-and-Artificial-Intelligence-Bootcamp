# Python Loops Challenges
Solve the following challenges to practice `for`, `while`, `range`, loop control statements, and nested loops.

**Challenge 1:** Print numbers from 1 to 10 using a `for` loop.

````python
#write your solution

for i in range(10):
    print(i+1)
````

**Challenge 2:** Print even numbers from 2 to 20 using `for` and `range()`.

````python
#write your solution

for i in range(1, 21):
    if i %2 ==0:
        print(i) 
````

**Challenge 3:** Print numbers from 5 to 1 using a `while` loop.

````python
#write your solution
i = 5
while i > 0:
    print(i)
    i -= 1
````

**Challenge 4:** Loop through the string `'Python'` and print each letter.

````python
#write your solution

s = "python"
for i in s:
    print(i)
````

**Challenge 5:** Print the square of numbers from 1 to 5.

````python
#write your solution
for i in range (1,6):
    print(i**2)
````

**Challenge 6:** Use a `for` loop to calculate the sum of numbers from 1 to 100.

````python
#write your solution
sum = 0
for i in range(1,101):
    sum += i

print(sum)
````

**Challenge 7:** Use `break` to stop the loop when the number reaches 5 (from 1 to 10).

````python
#write your solution

for i in range(1,11):
    if i == 5:
        break
    print(i)
````

**Challenge 8:** Use `continue` to skip printing number 3 (from 1 to 5).

````python
#write your solution

for i in range(1,6):
    if i == 3:
        continue
    print(i)
````

**Challenge 9:** Use `pass` inside a loop that iterates over a list of fruits.

````python
#write your solution
fruits = ["banana","apple","orange"]

for i in fruits:
    pass
````

**Challenge 10:** Use nested loops to print pairs of numbers from 1 to 3 (i, j).

````python
#write your solution
for i in range(1,4):
    for j in range(1,4):
        print(i,"", j)
````

**Challenge 11:** Print this pattern using nested loops:
```
*
**
***
```

````python
#write your solution

for i in range(1, 4):
    for j in range(i):
        print("*", end="")
    print()
````

**Challenge 12:** Print a multiplication table for 1 to 3 using nested loops.

````python
#write your solution

for i in range(1,4):
    for j in range(1,4):
        print(f"{i} * {j} = {i*j}")
````

**Challenge 13:** Count how many even numbers are between 1 and 20 using a loop.

````python
#write your solution
c = 0
for i in range(1,21):
    if i %2 ==0:
        c +=1
print(c)
````

**Challenge 14:** Iterate over this list and print only names longer than 3 letters:
```python
names = ['Ali', 'Sara', 'Noor', 'Om']```

````python
#write your solution
names = ['Ali', 'Sara', 'Noor', 'Om']

for n in names:
    if len(n) > 3:
        print(n)
````
