```python
i = 0
while i < 3:
    j = 1
    while j < 4:
        print(j)
        j += 1
    i += 1 
    print()
```

when `for loop` have `else`:

the `else` stetment don't happened until loop finish in normal way, if `break` else statement will not happen

```python
for i in range(2,10):
    for j in range(2,i):
        if i % j ==0:
            break
    else:
        print(i)
```

```python
for i in range(3):
    for j in range(i):
        print("*", end="")
    print()
```

```python
for i in range(1,3):
    for j in range(1,3):
        print(j, end=" ")
    print()
```

```python
for i in range(1,4):
    for j in range(1,4):
        print(f"{i}*{j} = {i*j}")
```
